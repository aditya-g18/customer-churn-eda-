```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df[['TotalCharges']]
```

```python
df[['TotalCharges']].isnull()
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].mean())
df['TotalCharges'].isnull().sum()
```

```python
df[['SeniorCitizen']].head(5)
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].replace({0: 'No', 1: 'Yes'})
df[['SeniorCitizen']].head(5)
```

```python
df['customerID'].duplicated().sum()
```

```python
ax = sns.countplot(x='Churn', data=df, color='lightblue', edgecolor='blue')
plt.title('Churn')
ax.bar_label(ax.containers[0])

plt.show()
```

```python
gd = df.groupby('Churn').agg({'Churn' : 'count'})
gd
plt.pie(gd['Churn'], labels=gd.index, autopct='%1.1f%%', colors=['lightgreen', 'green'])
plt.title('churn_percentage')
plt.show()
```

```python
plt.figure(figsize = (4,4))
w = sns.countplot(x='gender', data=df, hue='Churn')
for i in w.containers:
    w.bar_label(i)

plt.show()
```

```python
plt.figure(figsize = (4,4))

p = sns.countplot(x='SeniorCitizen', data=df, hue='Churn')
for n in p.containers:
    p.bar_label(n)
```

```python
plt.figure(figsize=(4,4))

ct = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

ax = ct.plot(kind='bar', stacked=True)

for c in ax.containers:
    ax.bar_label(c, fmt='%.1f%%')

plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage')
plt.title('SeniorCitizen vs Churn (%)')

plt.show()
```
